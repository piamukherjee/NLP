#### Problem definition

In this example we build a Content-based recommendation Engine to find similar books to books we have liked. 

Similarity here implies that the recommended books are similar to your favourite books in terms of their book descriptions as obtained from goodreads.com

Steps:

- First we learn to webscrape from goodreads.com to create a database of books with titles, authors, ratings and book descriptions. 

- Once we have a clean data frame we will do some exploratory data analysis to get acquainted with the data: make plots to look at the distribution of a) book ratings b) word length of book descriptions etc.

- We will then use tools that we discussed in the NLP primer course a) we first vectorize our book descriptions. To do this we can include all words and puctuation as they appear or we can remove stop words; we can also use TF-IDF weighting, which stands for term frequency–inverse document frequency. This weighting is used to evaluate how important a word is to a document in a collection or corpus. The importance of a word increases proportionally to the number of times a word appears in the document, but if the word is common across all documents in the corpus, then it gets downweighted. tf–idf is one of the most popular term-weighting schemes used for text summarization and in turn by search engines as a central tool in scoring and ranking a document's relevance given a user query. 

<tr>
<td><img src="TF-IDF.png" width="300" height="200" alt="">
<tr>

Term frequency stands for how often a word appears in a given document. Inverse document frequency downweights a word if it is common across documents of the corpus. Repeated words get a higher tf score. Rear words get a higher idf score. tf-idf = tf*idf.  
    
Example:
Consider a document containing 100 words wherein the word cat appears 3 times.
The term frequency (i.e., tf) for cat is then (3 / 100) = 0.03. Now, assume we have 10 million documents and the word cat appears in one thousand of these. Then, the inverse document frequency (i.e., idf) is calculated as log(10,000,000 / 1,000) = 4. Thus, the Tf-idf weight is the product of these quantities: 0.03 * 4 = 0.12.
    
Therefore we convert each book description into a TF-IDF vector. We can just use a software package - here TfidfVectorizer from sklearn - to convert a set of documents into TF-IDF vectors. The resulting TF-IDF matrix is sparse - each book description contains about 100 words, but the whole set together contains about 5000 words, so each book description maps onto a 5000 element vector that indicates whether or not it contains the corresponding word. Each column of the TF-IDF matrix corresponds to a feature or a word, each row is document, the whole matrix is sparse.
    
Text Analysis is a major application field for machine learning algorithms. However algorithms need numbers, so we go from a matrix of words to a matrux of numbers. Fortunately all this is automated for us. 

- We can now define a measure of similarity between two documents (here two book descriptions) using their TF-IDF vectors. A common way to do this is to use cosine similarity, which simply finds the cos of the angle between two vectors. If the vectors are the same, the angle will be 0 and the cosine of it will be 1. So 1 indicates perfect similarity.

We can now implement the algorithm and find the books that are most similar to a given book. 

- We can implement another method to do this and compare our results. The aim is to introduce the idea behind another important NLP technique. word2vec is an algorithm that convert a word into a vector by training a neural network on a very large corpus of words with the aim of capturing a lot of meaning about that word - its synonyms, the context in which it appears, what words usually follow or preceed it (syntactical information) etc. Google allows us to use the result of their word2vec algorithm which would have taken ages and a lot of processing power to run, but can now be used repeatedly for various tasks (this is called transfer learning).

For the second method we consider the word2vec representation of each word (specifying the level of detail we want ie. how long should each vector be, how much context (from neighbouring words) would we like to keep. We weight this by the td-idf score of each word to find a final vector representation for each book description by averaging the word2vec representation of each word over all words with tf-idf scores as weights:
$\frac{tf-idf(w1)*W2V(w1)+tf-idf(w2)*W2V(w2)+ ...}{tf-idf(w1)+tf_idf(w2)+.....}$
where w1, w2 etc are words of a given book description. 

Finally we compare vector representations of different documents (here book descriptions) via a similarity score. This is how we find similar books.

Remember: Word is not something that the computer will naturally understand. For a computer to understand the meaning behind a word, the word needs to be encoded in a numeric form. This is where the technique of word embeddings comes in. If two words have a similar context, like ‘tea’ and ‘coffee’, then the distance between these two words in embedding space will small.

word2vec is a powerful way to generate word embeddings - it converts each word into a vector which captures the latent information within the word. Since each document is a collection of words we can use word2vec to represent a document, say by averaging over its words. 

The tf-idf weighted word2vec representation we have used additionally takes into account the frequencies of the words - downweighting very common words. 

It is word vectors that make technologies such as speech recognition and machine translation possible. All advanced NLP tasks benefit from it. 
    
In our simpler recommendation system for movies, we simply used bag of words to indicate the presence or absence of a word in a document. That didn't capture all the information about a word's meaning or context. Relationships between words are missed - for instance dog and cat are often discussed in the context of pets or tea and coffee are drinks. Such words will have similar word vectors. Each numnber in the word's vector can be thought of as capturing a dimension or an aspect of the word's meaning (although they dont actually hold clearly defined meanings in a sense that we can understand, they are the weights of a neural network model).

```python
def getdata(base_site):

    # Make http request
    #response = requests.get(base_site)
    print(base_site)
    s = requests.Session()
    response = s.get(base_site)

    # Check if request is successful. Status code of 200 indicates a successful attempt.
    response.status_code

    # Get the html from webpage
    html = response.content

    # Creating a BeautifulSoup object with the use of a parser
    soup = BeautifulSoup(html, "lxml")

    # Exporting html file
    #with open('popularrunningbooks.html', 'wb') as file:
    #    file.write(soup.prettify('utf-8'))
    #with open('bestbooks.html', 'wb') as file:
    #    file.write(soup.prettify('utf-8'))
    
    #### Extracting the url addresses of each book link ####

    # First layer: The element that contains all the data
    divs = soup.find_all("div", {"class": "elementList"})

    # Second layer: Extracting html tags that contain the links
    links = [div.find('a') for div in divs]

    # Extracting the partial links  
    relative_url = [link['href'] for link in links]  

    # Computing the full url addresses 
    full_url = [urljoin(base_site, relativeurl) for relativeurl in relative_url]
    
    # Filter only the book links
    book_url = [url for url in full_url if "https://www.goodreads.com/book/show" in url]

    #### Scraping information of each book using for loop ####
    df = pd.DataFrame()
    book_description = []
    book_author = []
    book_title = []
    book_rating = []
    book_pages = []

    #creating a loop counter
    i = 0

    #Loop through all 50 books
    for url in book_url:
    
        #connect to url page
        note_resp = requests.get(url)
    
        #checking if the request is successful
        if note_resp.status_code == 200:
            print("URL{}: {}".format(i+1, url))
    
        else:
            print('Status code{}: Skipping URL #{}: {}'.format(note_resp.status_code, i+1, url))
            i = i+1
            continue
    
        #get HTML from url page
        note_html = note_resp.content
    
        #create beautifulsoup object for url page
        note_soup = BeautifulSoup(note_html, 'html.parser')
    
        #Extract Author particulars
        author_divs = note_soup.find_all("div", {"class":"authorName__container"})
        author_text = author_divs[0].find_all('a')[0].text
        book_author.append(author_text)
    
        #Extract title particulars
        title_divs = note_soup.find_all("div", {"class": "last col"})
        title_text = title_divs[0].find_all('h1')[0].text
        book_title.append(title_text)
    
        #Extract rating particulars
        rating_divs = note_soup.find_all("div", {"class": "uitext stacked", "id": "bookMeta"})
        rating_text = rating_divs[0].find_all("span", {"itemprop": "ratingValue"})[0].text
        book_rating.append(rating_text)
    
        #Extracting page particulars
        page_divs = note_soup.find_all("div", {"class": "row"})
        try:
            page_text = page_divs[0].find_all("span", {"itemprop": "numberOfPages"})[0].text.strip(' pages')
        except IndexError:
            page_text = 0
        book_pages.append(page_text)
    
        #Extracting description particulars
        description_divs = note_soup.find_all("div", {"class": "readable stacked", "id": "description"})
        try:
            description_text = description_divs[0].find_all("span")[1].text
        except IndexError:
            try:
                description_text = description_divs[0].find_all("span")[0].text
            except IndexError:
                description_text = "Nil"
        book_description.append(description_text)
        
        #Incremeting the loop counter
        i = i+1
    
    #### Some simple data processing ####

    revised_book_title = [book.strip() for book in book_title]
    revised_book_rating = [float(rating.strip()) for rating in book_rating]
    revised_book_pages = [int(page) for page in book_pages]
    revised_book_description = [description.strip() for description in book_description]
    
    df["Book Title"]=revised_book_title
    df["Author"]=book_author
    df["Rating"]=revised_book_rating
    df["Pages"]=revised_book_pages
    df["Description"]=revised_book_description
    df["Links"]=book_url

    return df
```

If you want to improve a sentiment analysis model or create your own you could input what you think is the more accurate sentiment by hand. You have a good understanding of language and you might think you can do better than TextBlob at summing up the sentiment of each book description. Once you have sentiment labels to go with your inputs you can train a machine learning model to learn the link between the inputs and labels. There are many machine learning models to choose from including neural networks and deep learning models that will find different ways to link the labels to the inputs. Once the model has been trained it can be used to assign labels to new inputs. In this way, without having to specify detailed rules, the machine 'learns' the relationship between the inputs (these could be text, audio, images) and the labels. Hence 'machine learning'. Before it is used on unseen examples it is usually tested on a set of inputs and labels that you kept aside from the initial set, for accuracy.  
In this way one can build a machine learning method for sentiment analysis rather than a rule based method which relies on lexicon of words that have a sentiment assigned to them, as above. Can you see the drawback of using a rule based method - it is faster, easier, but doesn't include meaning, context or intuition about what is actually positive os negative sentiment.  

Should we try to build a machine learning model? 

```python
#from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df['cleaned'],df['Rating'],test_size=0.2)
```

Since our X is in text form we need to first find an appropriate numeric representation for them as we had done before. Then we'll get to building a ML model.  

```python
# Exactly what we had done before but this time applied to x_train and x_test separately.
# Although in order to have tfidf info for all words we are doing this on the whole corpus. 
# If we abandon the tfidf part then we can do this for just x_train and never really see x_test.

#splitting the description into words

corpus = []
for words in df['cleaned']:
    corpus.append(words.split())
    
google_model.intersect_word2vec_format(EMBEDDING_FILE, lockf=1.0, binary=True)
google_model = Word2Vec(size = 300, window=5, min_count = 2, workers = -1)
google_model.build_vocab(corpus)
google_model.train(corpus,total_examples=google_model.corpus_count, epochs = 5)


tfidf = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df = 5, stop_words='english')
tfidf.fit(df['cleaned'])
tfidf_list = dict(zip(tfidf.get_feature_names(), list(tfidf.idf_)))
tfidf_feature = tfidf.get_feature_names() # tfidf words/col-names
```

```python
# Building TF-IDF Word2Vec 
def build_tfidfWordVector(text,tfidf_list):
# Storing the TFIDF Word2Vec embeddings
    tfidfword2vec_vectors = []; 
    line = 0;
# for each book description
    for desc in text: 
  # Word vectors are of zero length (Used 300 dimensions)
        sent_vec = np.zeros(300) 
    # num of words with a valid vector in the book description
        weight_sum =0; 
    # for each word in the book description
        for word in desc.split(): 
            if word in google_model.wv.vocab and word in tfidf_feature:
                vec = google_model.wv[word]
                tf_idf = tfidf_list[word] * (desc.count(word) / len(desc))
                sent_vec += (vec * tf_idf)
                weight_sum += tf_idf
        if weight_sum != 0:
            sent_vec /= weight_sum
        tfidfword2vec_vectors.append(sent_vec)
        line += 1
    return tfidfword2vec_vectors
```

```python
from sklearn.preprocessing import scale
train_vecs=build_tfidfWordVector(x_train,tfidf_list)
train_vecs=scale(train_vecs)

test_vecs=build_tfidfWordVector(x_test,tfidf_list)
test_vecs=scale(test_vecs)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
reg = LinearRegression()
reg.fit(train_vecs,y_train)
y_pred=reg.predict(test_vecs)
print(reg.score(test_vecs,y_test))
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error=', rmse)
```
